# Task 2
Extract the following web URL text using BeautifulSoup:
https://en.wikipedia.org/wiki/Google

```python
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/Google"
page = requests.get(url)
parse = BeautifulSoup(page.text, "html.parser")
p = parse.findAll('p')
```

# Task 3
Save it in input.txt

```python
f = open("input.txt", "w")
f.writelines([str(item.text) for item in p])
f.close()
```

# Task 4
Apply the following on the “input.txt”and show output:
a. Tokenization
b. POS
c. Stemming
d. Lemmatization
e. Trigram
f. Named Entity Recognition

## Task a
### Tokenization

```python
from nltk.tokenize import word_tokenize
f = open('input.txt')
sentences = f.readlines()
sentence = sentences[1]
sentences[:4]
```

```python
sentence = sentences[3]
tokens = word_tokenize(sentence)
tokens
```

## Task b
### POS Tagging

```python
from nltk import pos_tag
tagged_sentence = pos_tag(tokens)
tagged_sentence
```

# Task c
### Stemming

```python
# PorterStemmer
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = sentence.split(' ')
[stemmer.stem(word) for word in words]
```

```python
# LancasterStemmer
from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()

[lancaster_stemmer.stem(word) for word in words]
```

```python
# SnowballStemmer
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')

[snowball_stemmer.stem(word) for word in words]
```

# Task d
### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatiser = WordNetLemmatizer()
[lemmatiser.lemmatize(word) for word in words]
```

# Task e
### Trigrams

```python
from nltk import trigrams
trigram_list = list(trigrams(tokens))
trigram_list
```

# Task f
### Named Entity Recognition

```python
from nltk import ne_chunk
print(ne_chunk(tagged_sentence, binary=True))
```
