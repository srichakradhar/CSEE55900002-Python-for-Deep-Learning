```python
import pandas
from keras.models import Sequential
from keras.layers.core import Dense, Activation

from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
# load dataset
dataset = pd.read_csv("diabetes.csv", header=None).values

dataset.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=8, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
my_first_nn.summary()
```

```python
my_first_nn.evaluate(X_test, Y_test)
```

### Add more Dense layers to the existing code and check how the accuracy changes.

```python
dense_nn = Sequential() # create model
dense_nn.add(Dense(60, input_dim=8, activation='relu')) # hidden layer
dense_nn.add(Dense(80, activation='relu')) # hidden layer
dense_nn.add(Dense(1, activation='sigmoid')) # output layer
dense_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
dense_nn_fitted = dense_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
dense_nn.summary()
```

```python
dense_nn.evaluate(X_test, Y_test)
```

### Change the data source to Breast Cancer dataset * available in the source folder and make required changes

```python
breast_cancer_df = pd.read_csv("breastcancer.csv", header=[0])

breast_cancer_df.head()
```

```python
breast_cancer_df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
breast_cancer_df.columns
```

```python
breast_cancer_df['diagnosis'].replace(['M','B'], [1, 0], inplace=True)
breast_cancer_df.head()
```

```python
breast_cancer_data = breast_cancer_df.values
breast_cancer_data.shape
```

```python
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(breast_cancer_data[:,1:31], breast_cancer_data[:,0],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
breast_cancer_nn = Sequential() # create model
breast_cancer_nn.add(Dense(120, input_dim=30, activation='relu')) # hidden layer
dense_nn.add(Dense(80, activation='relu')) # hidden layer
breast_cancer_nn.add(Dense(1, activation='sigmoid')) # output layer
breast_cancer_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
breast_cancer_nn_fitted = breast_cancer_nn.fit(X_train1, Y_train1, epochs=100,
                                     initial_epoch=0)
breast_cancer_nn.summary()
```

```python
breast_cancer_nn.evaluate(X_test1, Y_test1)
```
